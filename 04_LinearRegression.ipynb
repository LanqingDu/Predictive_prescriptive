## [2. Simple Linear Regression:](#howCoding1)

```r
# ================================
# Step-1: Load Required Package
# ================================

library(ggplot2)
library(dplyr)
library(readxl)
```

```r
# ================================
# Step-3: Read in data file
# ================================

df <- read_excel("ButlerWithDeliveries.xlsx")
head(df)

# show to column names to avoid any possible typos
print(colnames(df))
```

```r
# ================================
# Step-4: Scatter plot
# ================================

scatter1 <- ggplot(df, aes(x = Miles, y = Time)) + geom_point()
print(scatter1)
```

```r
# ================================
# Step-5: Linear regression
# ================================

# library(stats)

# lm(formula, data)
#   formula: a symbolic description of the model (e.g., response ~ predictor(s)).
#   data: a data frame that contains the variables used in the model.
#   return: an object of class 'lm' that represents the fitted linear regression model.
#   model_object: stores the resulting linear regression fit.
#   output: a model object that can be used for evaluating coefficients, goodness-of-fit, diagnostics, and prediction.

# slm <- lm(Time ~ Miles, data = df)
#   formula: Time ~ Miles  models 'Time' as a linear function of 'Miles'.
#   data: df  the data frame containing the 'Time' and 'Miles' variables.
#   return: a linear model object containing estimates for intercept and slope.
#   slm: stores the fitted simple linear regression model for predicting 'Time' based on 'Miles'.
#   output: a model object used for summary analysis, residual checks, and future predictions.

slm <- lm(Time ~ Miles, data = df)
print(slm)
```

```r
# ================================
# Step-6: Assess the fit
# ================================

# summary(object)
#   object: a fitted model object (e.g., of class 'lm' from a linear regression).
#   return: a detailed summary of the fitted model including coefficient estimates,
#           standard errors, t-values, p-values, residual statistics, and goodness-of-fit metrics.
#   summary_object: stores the output summary information (usually printed directly).
#   output: comprehensive model diagnostics and inference results useful for understanding
#           the relationship between predictors and the response.


# summary(slm)
#   object: slm  the linear regression model fitted by lm(Time ~ Miles, data = df).
#   return: a summary of the model including estimates for intercept and slope,
#           statistical significance of predictors, residual standard error, R-squared, etc.
#   summary_object: stores the detailed summary of the simple linear regression.
#   output: printed report used to assess model fit, interpret coefficients,
#           and check significance of 'Miles' in predicting 'Time'.
summary(slm)
```

```r
# ================================
# Step-7: Make predictions
# ================================

# predict(object, newdata)
#   object: a fitted model object (e.g., an 'lm' object from linear regression).
#   newdata: a data frame containing new values of predictor variables for which predictions are desired.
#   return: predicted values of the response variable based on the model and new predictor values.
#   predictions: stores the predicted response values.
#   output: numeric vector or data frame of predicted responses corresponding to newdata inputs.

# y_hat <- predict(slm, newdata = data.frame(Miles = 80))
#   object: slm  the linear regression model fitted by lm(Time ~ Miles, data = df).
#   newdata: a data frame with one row where 'Miles' equals 80.
#   return: predicted 'Time' value for when 'Miles' is 80.
#   y_hat: stores the predicted Time from the model.
#   output: numeric value representing the estimated Time for 80 Miles based on the fitted model.
y_hat <- predict(slm, data.frame(Miles = 80))
print(y_hat)
```

## [3. Multiple Linear Regression:](#howCoding2)

```r
# Load required libraries

library(readxl)

# Read the Excel file
df <- read_excel("ButlerWithDeliveries.xlsx")

# Preview the data
head(df)
print(colnames(df))

# Build a multiple linear regression model

# lm(formula, data)
#   formula: a symbolic description of the model, where the response variable
#            is modeled as a function of one or more predictor variables.
#   data: a data frame containing the variables referenced in the formula.
#   return: an object of class 'lm' representing the fitted linear regression model.
#   model_object: stores the resulting model fit including coefficients for all predictors.
#   output: a model object used for interpretation, diagnostics, and prediction.

# mlr <- lm(Time ~ Miles + Deliveries, data = df)
#   formula: Time ~ Miles + Deliveries  models 'Time' as a linear function of both 'Miles' and 'Deliveries'.
#   data: df  the data frame containing variables 'Time', 'Miles', and 'Deliveries'.
#   return: a linear model object estimating coefficients for intercept, 'Miles', and 'Deliveries'.
#   mlr: stores the multiple linear regression model fit.
#   output: a model object used to analyze how both 'Miles' and 'Deliveries' affect 'Time',
#           and to make predictions or perform diagnostics.
mlr <- lm(Time ~ Miles + Deliveries, data = df)


# residual plot

# function: plot(x, y, ...)  creates a scatter plot of 'y' versus 'x'.
#   x: df$Miles  the predictor variable 'Miles' from the data frame 'df'.
#   y: mlr$residuals  residuals from the multiple linear regression model 'mlr'.
#   xlab: "Miles"  label for the x-axis.
#   ylab: "Residuals"  label for the y-axis.
#   main: "Residual Plot"  title of the plot.
#   return: a plot object (rendered immediately, not stored by default).
#   output: a residual plot visualizing the relationship between 'Miles' and the model residuals,
#           used to check linearity, constant variance (homoscedasticity), and independence assumptions.
plot(df$Miles, mlr$residuals,
     xlab = "Miles",
     ylab = "Residuals",
     main = "Residual Plot")

# function: abline(...)  adds a straight line to an existing plot.
#   h: 0  draws a horizontal line at y = 0.
#   col: "red"  sets the color of the line to red.
#   lty: 2  sets the line type to dashed.
#   return: adds a line to the current plot; does not produce a separate object.
#   output: a horizontal dashed red line at y = 0 on the existing plot,
#           commonly used as a reference line for residuals in a residual plot.
abline(h = 0, col = "red", lty = 2)



# Display the model summary
summary(mlr)

# Predict Time when Miles = 80 and Deliveries = 30

# data.frame(...)
#   ...: named arguments representing variables and their values to create a new data frame.
#   return: a data frame constructed from the provided vectors or values.
#   new_data: stores the data frame for use in functions like predict().
#   output: a tabular data structure with specified variables and values, matching the models predictors.

# new_data <- data.frame(Miles = 80, Deliveries = 30)
#   Miles = 80, Deliveries = 30: specify values for the predictor variables used in the model.
#   return: a data frame with one row and two columns named 'Miles' and 'Deliveries'.
#   new_data: stores this data frame, which can be passed to predict() to estimate the response.
#   output: a one-row data frame representing a scenario for prediction of 'Time' given these inputs.
new_data <- data.frame(Miles = 80, Deliveries = 30)

y_hat <- predict(mlr, newdata = new_data)

# Print the predicted value
cat("y_hat =", y_hat, "\n")
```

```r
# clean version

library(readxl)

# Read the Excel file
df <- read_excel("ButlerWithDeliveries.xlsx")

# Preview the data
head(df)
print(colnames(df))

# fit model
mlr <- lm(Time ~ Miles + Deliveries, data = df)


# residual plot
plot(df$Miles, mlr$residuals,
     xlab = "Miles",
     ylab = "Residuals",
     main = "Residual Plot")
abline(h = 0, col = "red", lty = 2)



# Display the model summary
summary(mlr)

# Predict Time when Miles = 80 and Deliveries = 30
new_data <- data.frame(Miles = 80, Deliveries = 30)

y_hat <- predict(mlr, newdata = new_data)

# Print the predicted value
cat("y_hat =", y_hat, "\n")
```
